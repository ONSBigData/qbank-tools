```python
def matches(index, smatrix=csm, top=5, thresh=0.5):
    scores = smatrix[index].copy()

    HIGH_VAL = 999
    own_score = scores[index]
    scores[index] = HIGH_VAL

    # put this index on first place
    indices = scores.argsort()[::-1]

    scores.sort()
    scores = scores[::-1]

    comb = list((i, s if s < HIGH_VAL else own_score) for i, s in zip(indices, scores) if s >= thresh)
    if top != None:
        comb = comb[:top]
        
    mdf = df.loc[[i for i, _ in comb]]
    mdf['score'] = [s for _, s in comb]
    return mdf

matches(30)
```

```python
ms = []

for i in range(1000):
    m = matches(i)
    m = m[(m['survey'] != m['survey'][i]) | (m.index == i)]
    m = m[m['score'] > 0.75]
    
    if len(m) > 1:
        ms.append(m)
        
pd.concat(ms).to_csv('examples.csv')
```
